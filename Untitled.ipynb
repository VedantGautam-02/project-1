```python
# Cell 1 — Import Libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Display settings
pd.set_option('display.max_columns', None)

# Style
sns.set_style("darkgrid")

print("Libraries Imported Successfully")
```

```python
# Cell 2 — Load Dataset

df = pd.read_csv('/Users/zeeshanhussain/Desktop/DA proect/45.csv')

print("Dataset Loaded Successfully")
```

```python
# Cell 3 — View Dataset

df.head()
```

```python
# Cell 5 — Check Missing Values

df.isnull().sum()
```

```python
# Cell 7 — Basic Statistics

df.describe()
```

```python
# Cell 8 — Create Engagement Rate

df['Engagement Rate'] = (df['Clicks'] / df['Impressions']) * 100

df.head()
```

```python
# Create Total Conversions

df['Total Conversions'] = df['Clicks'] * (df['Conversion_Rate'] / 100)
```

```python
# Fill Missing Values

df['Total Conversions'] = df['Total Conversions'].fillna(0)

df['Cost Per Click (CPC)'] = df['Cost Per Click (CPC)'].fillna(0)

df['Cost Per Acquisition'] = df['Cost Per Acquisition'].fillna(0)

print("Missing Values Fixed")
```

```python
# Convert Columns to Numeric

df['Acquisition_Cost'] = pd.to_numeric(
    df['Acquisition_Cost'],
    errors='coerce'
)

df['Cost Per Click (CPC)'] = pd.to_numeric(
    df['Cost Per Click (CPC)'],
    errors='coerce'
)

df['Cost Per Acquisition'] = pd.to_numeric(
    df['Cost Per Acquisition'],
    errors='coerce'
)

df['Total Conversions'] = pd.to_numeric(
    df['Total Conversions'],
    errors='coerce'
)

print("Columns Converted Successfully")
```

```python
# KPI Calculations

average_conversion_rate = df['Conversion_Rate'].mean()

average_engagement_rate = df['Engagement Rate'].mean()

total_acquisition_cost = df['Acquisition_Cost'].sum()

average_cpc = df['Cost Per Click (CPC)'].mean()

average_cpa = df['Cost Per Acquisition'].mean()

total_conversions = df['Total Conversions'].sum()

# Print KPIs

print("Average Conversion Rate :", round(average_conversion_rate, 2))

print("Average Engagement Rate :", round(average_engagement_rate, 2))

print("Total Acquisition Cost :", round(total_acquisition_cost, 2))

print("Average CPC :", round(average_cpc, 2))

print("Average CPA :", round(average_cpa, 2))

print("Total Conversions :", round(total_conversions, 2))
```

```python
# Campaign Type Analysis

campaign_analysis = df.groupby('Campaign_Type')[[
    'ROI',
    'Conversion_Rate',
    'Engagement Rate'
]].mean().sort_values(by='ROI', ascending=False)

campaign_analysis
```

```python
# ROI by Campaign Type

plt.figure(figsize=(10,6))

sns.barplot(
    x=campaign_analysis.index,
    y=campaign_analysis['ROI']
)

plt.title("ROI by Campaign Type")

plt.xticks(rotation=45)

plt.show()
```

```python
# Channel Analysis

channel_analysis = df.groupby('Channel_Used')[[
    'ROI',
    'Conversion_Rate'
]].mean().sort_values(by='ROI', ascending=False)

channel_analysis
```

```python
# Channel ROI Visualization

plt.figure(figsize=(10,6))

sns.barplot(
    x=channel_analysis.index,
    y=channel_analysis['ROI']
)

plt.title("ROI by Marketing Channel")

plt.xticks(rotation=45)

plt.show()
```

```python
# ROI Distribution by Campaign Type

plt.figure(figsize=(12,6))

sns.boxplot(
    x='Campaign_Type',
    y='ROI',
    data=df
)

plt.title("ROI Distribution by Campaign Type")

plt.xticks(rotation=45)

plt.show()
```

```python
# Violin Plot

plt.figure(figsize=(12,6))

sns.violinplot(
    x='Campaign_Type',
    y='ROI',
    data=df
)

plt.title("ROI Density by Campaign Type")

plt.xticks(rotation=45)

plt.show()
```

```python
# Boxplot + Data Points

plt.figure(figsize=(12,6))

sns.boxplot(
    x='Campaign_Type',
    y='ROI',
    data=df
)

sns.stripplot(
    x='Campaign_Type',
    y='ROI',
    data=df,
    color='red',
    alpha=0.4
)

plt.title("ROI Spread with Data Points")

plt.xticks(rotation=45)

plt.show()
```

```python
# Heatmap Analysis

pivot_table = df.pivot_table(
    values='ROI',
    index='Campaign_Type',
    columns='Channel_Used',
    aggfunc='mean'
)

plt.figure(figsize=(12,6))

sns.heatmap(
    pivot_table,
    annot=True,
    cmap='coolwarm'
)

plt.title("ROI Heatmap")

plt.show()
```
